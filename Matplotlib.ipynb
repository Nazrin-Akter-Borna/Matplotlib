### Axes Lavels, Legend and Grid

```python
days = [1,2,3,4,5,6]
max_t = [50,51,52,53,54,48]
min_t = [42,41,43,40,38,33]
avg_t = [42,41,43,40,35,33]

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend(loc = "best",shadow=True)
plt.grid()
```

## Bar Chart

```python
import numpy as np
```

```python
product = np.array([1,2,3,4])
price = [15,25,40,30]
profit = [5,7,10,5]

plt.bar(product-.2,price, label='price',color='red',width=.4)
plt.bar(product+.3,profit, label='profit',width=.4)
plt.legend()

plt.savefig("barchart.png",transparent = True, pad_inches = 2)
```

# Pie chart

```python
val = [300,500,1600,450,360]
lab = ['Pice','sell','Total','Shirt','Shari']
```

```python
plt.pie(val, labels = lab,autopct='%.2f%%',shadow=True,startangle=45,explode = [0,.2,0,.1,0])

plt.savefig("piechart.png",transparent = True, pad_inches = 2)
```
